```python
import os
#returns the absolute path of the working directory where the python is currently running
cwd = os.getcwd()
```

```python
fileList = os.listdir(cwd)
print(fileList)
```

```python
def ngram_text(n):
  for file in fileList:
    with open(cwd + "/"+file,'r') as f:
      text = str(f.read())
      textL = list(text)
      n_gram= ngrams(textL,n)
      x = [''.join(n) for n in n_gram]
      print(x)
      #return(x)
```

```python
ngram_text(3)
```

```python
def jaccard(x,y):
  a=x.intersection(y)
  b=x.union(y)
  return len(a)/len(b)
```

```python
doc1 = open(os.getcwd()+'/D1.txt','r')
doc2 = open(os.getcwd()+'/D2.txt','r')
```

```python
sentence_1 = list(doc1.read())
sentence_2 = list(doc2.read())
```

```python
n_gram = ngrams(sentence_1,3)
x = [''.join(n) for n in n_gram]
print(x)
```

```python
n_gram = ngrams(sentence_2,3)
y = [''.join(n) for n in n_gram]
print(y)
```

```python
print(jaccard(set(x),set(y)))
```
