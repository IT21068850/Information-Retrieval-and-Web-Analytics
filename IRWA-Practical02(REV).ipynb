```python
#i) schizophrenia AND drug
#AND operator --> intersection

#lets create a function for AND operator

def AND_op(post1,post2):
  if (post1 and post2):# check whether its empty.(if both are empty it returns false)
    return set(post1).intersection(set(post2))

#OR operator --> union
def OR_op(post1,post2):
  if (post1 and post2):
    return set(post1).union(set(post2))

#NOT operator --> difference
def NOT_op(post1,post2):
  if (post1 and post2):
    return set(post1).difference(set(post2))
```

```python
#(I)
for key in wordsAdded:
  if key == "schizophrenia":
    l1 = wordsAdded[key]
  if key == "drug":
    l2= wordsAdded[key]

AND_op(l1,l2)
```

```python
#(II)

for key in wordsAdded:
  if key == "for":
    l1 = wordsAdded[key]
  if key == "drug":
    l2= wordsAdded[key]
  if key == "approach":
    l3= wordsAdded[key]

AND_op(l1,NOT_op(l1,OR_op(l2,l3)))
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
quote = "Pythoners are very intelligent and work very pythonly and now they are pythoning their way to success."
```

```python
#First tokenize the words
tknwords = word_tokenize(quote)
tknwords
```

```python
# stop words in english
stopwords = set(stopwords.words('english'))
```

```python
filtered_list = []
# turn it to lower case

for word in tknwords:
  if word.casefold() not in stopwords:
    filtered_list.append(word)


print(filtered_list)
```

```python
# if i want to add my own stop words?
stp_new = list(stopwords)
stp_new.extend(['intelligent', 'work'])
```

```python
#new filteretd list
new_filtered_list = []
# turn it to lower case

for word in tknwords:
  if word.casefold() not in stp_new:
    new_filtered_list.append(word)


print(new_filtered_list)
```

```python
#Download the Libararies
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import re #regular expression
from nltk.tokenize import word_tokenize #to tokenize while cleaning the data
from nltk.corpus import stopwords #this is for removing stop words
from nltk.stem import PorterStemmer # for stemming
```

```python
stp = set(stopwords.words('english'))
dictionary2 = dict()
```

```python

for file in files:
  with open(cwd +'/'+ file,'r') as s:
   f =s.read().lower().replace('\n',' ')[0:]
   #cleaning the words
   f= re.sub(' ',' ',f)
   f= re.sub(r"s'","s",f)
   f = re.sub(r"\'ve","have",f)
   f = re.sub(r"'s","s",f)
   f = re.sub(r"\-"," ",f)
   f = re.sub(r"[^\w\s]"," ",f) # replaces all commas and full stops with a white space
   #print(f)

   # tokenizing the words
   words =  word_tokenize(f)
   #print(len(words))


   #removing stop words
   re_stp = [] #intializing reamining words after removal of stop words
   for word in words:
    if word not in stp:
      re_stp.append(word)


   # stemming
   stem_list = []
   ps = PorterStemmer()
   for i in re_stp:
    stem_list.append(ps.stem(i))
    #print(stem_list)



   #append the final cleaned tokens
   temp_dict= dict()
   a = 0

   for x in stem_list:
    key = x
    temp_dict.setdefault(key,[])
    temp_dict[key].append(a)
    a += 1

   # print(temp_dict)
   for x in temp_dict:
    if dictionary2.get(x):
      dictionary2[x][file] = temp_dict.get(x)

    else:
      key = x
      dictionary2.setdefault(key,[])
      dictionary2[key]={}
      dictionary2[x][file] = temp_dict.get(x)


print(dictionary2)







```
